````python
# 7. Trading Signals Analysis

# Calculate technical indicators
# RSI (Relative Strength Index)
def calculate_rsi(data, periods=14):
    delta = data.diff()
    gain = (delta.where(delta > 0, 0)).rolling(window=periods).mean()
    loss = (-delta.where(delta < 0, 0)).rolling(window=periods).mean()
    rs = gain / loss
    return 100 - (100 / (1 + rs))

df['RSI'] = calculate_rsi(df['close'])

# MACD (Moving Average Convergence Divergence)
df['EMA12'] = df['close'].ewm(span=12).mean()
df['EMA26'] = df['close'].ewm(span=26).mean()
df['MACD'] = df['EMA12'] - df['EMA26']
df['Signal_Line'] = df['MACD'].ewm(span=9).mean()

df.head()
df.shape
````

````python
# Convert date if available, otherwise generate index-based time features
if 'date' in df.columns:
    df['date'] = pd.to_datetime(df['date'])
else:
    df['date'] = pd.date_range(start='1/1/2014', periods=len(df), freq='D')  # Assuming daily data


# Time-Based Features
df['day_of_week'] = df['date'].dt.dayofweek  # Monday=0, Sunday=6
df['week_of_year'] = df['date'].dt.isocalendar().week
df['month'] = df['date'].dt.month
df['is_weekend'] = df['day_of_week'].isin([5, 6]).astype(int)

# Bollinger Bands
window = 20  # Typical period for Bollinger Bands
df['BB_middle'] = df['close'].rolling(window=window).mean()
df['BB_std'] = df['close'].rolling(window=window).std()
df['BB_upper'] = df['BB_middle'] + (2 * df['BB_std'])
df['BB_lower'] = df['BB_middle'] - (2 * df['BB_std'])
df['BB_width'] = df['BB_upper'] - df['BB_lower']

# Stochastic Oscillator
df['low_14'] = df['low'].rolling(window=14).min()
df['high_14'] = df['high'].rolling(window=14).max()
df['stoch_k'] = 100 * (df['close'] - df['low_14']) / (df['high_14'] - df['low_14'])
df['stoch_d'] = df['stoch_k'].rolling(window=3).mean()

# ATR (Average True Range)
df['high_low'] = df['high'] - df['low']
df['high_close_prev'] = np.abs(df['high'] - df['close'].shift(1))
df['low_close_prev'] = np.abs(df['low'] - df['close'].shift(1))
df['true_range'] = df[['high_low', 'high_close_prev', 'low_close_prev']].max(axis=1)
df['ATR'] = df['true_range'].rolling(window=14).mean()

# On-Balance Volume (OBV)
df['OBV'] = (np.sign(df['close'].diff()) * df['volume']).fillna(0).cumsum()

# VWAP (Volume-Weighted Average Price)
df['vwap'] = (df['volume'] * (df['high'] + df['low'] + df['close']) / 3).cumsum() / df['volume'].cumsum()

# Rolling Window Features
df['rolling_mean_7'] = df['close'].rolling(window=7).mean()
df['rolling_mean_30'] = df['close'].rolling(window=30).mean()
df['rolling_vol_7'] = df['close'].rolling(window=7).std()
df['rolling_vol_30'] = df['close'].rolling(window=30).std()
df['rolling_return_7'] = df['close'].pct_change(periods=7)
df['rolling_return_30'] = df['close'].pct_change(periods=30)
df['rolling_max_7'] = df['high'].rolling(window=7).max()
df['rolling_min_7'] = df['low'].rolling(window=7).min()

# Lag Features
df['close_lag_1'] = df['close'].shift(1)
df['close_lag_3'] = df['close'].shift(3)
df['close_lag_7'] = df['close'].shift(7)
df['close_lag_30'] = df['close'].shift(30)

# Derived Features
df['trend_strength'] = df['MA50'] / df['MA200']
df['volatility_volume_ratio'] = df['volatility'] / df['volume']
df['MACD_histogram'] = df['MACD'] - df['Signal_Line']

# Drop temporary columns used in calculations
df.drop(columns=['low_14', 'high_14', 'high_low', 'high_close_prev', 'low_close_prev', 'true_range'], inplace=True)

# Show updated dataset with new features
df.head()
````

````python
# First, let's inspect and clean the original data
def inspect_series(series, name):
    print(f"\nInspecting {name}:")
    print(f"First 5 values:\n{series.head()}")
    print(f"Basic stats:\n{series.describe()}")
    
# Reset calculations from scratch
def calculate_clean_returns(prices):
    # Calculate simple returns
    returns = prices.pct_change()
    # Remove any extreme values (e.g., +/- 100%)
    returns = returns.clip(lower=-1, upper=1)
    # Remove NaN and inf values
    returns = returns.replace([np.inf, -np.inf], np.nan).dropna()
    return returns

# Clean performance metrics calculation
def calculate_performance_metrics(prices, name=""):
    try:
        # Calculate clean returns
        returns = calculate_clean_returns(prices)
        
        # Calculate cumulative return
        cumulative_return = (1 + returns).cumprod().iloc[-1] - 1
        
        # Calculate annualized return
        n_years = len(returns) / 252
        annual_return = (1 + cumulative_return) ** (1/n_years) - 1
        
        # Calculate volatility
        volatility = returns.std() * np.sqrt(252)
        
        # Calculate Sharpe Ratio
        risk_free_rate = 0.02
        sharpe_ratio = (annual_return - risk_free_rate) / volatility
        
        # Calculate Maximum Drawdown
        cum_returns = (1 + returns).cumprod()
        rolling_max = cum_returns.expanding().max()
        drawdowns = (cum_returns - rolling_max) / rolling_max
        max_drawdown = drawdowns.min()
        
        metrics = pd.Series({
            'Total Return': cumulative_return,
            'Annual Return': annual_return,
            'Volatility': volatility,
            'Sharpe Ratio': sharpe_ratio,
            'Max Drawdown': max_drawdown
        })
        
        return metrics
    
    except Exception as e:
        print(f"Error in {name} calculations: {e}")
        return pd.Series({
            'Total Return': np.nan,
            'Annual Return': np.nan,
            'Volatility': np.nan,
            'Sharpe Ratio': np.nan,
            'Max Drawdown': np.nan
        })

# Inspect original price data
inspect_series(df['close'], 'Close Prices')

# Calculate market performance
market_metrics = calculate_performance_metrics(df['close'], "Market")

# Calculate strategy performance
# First, calculate strategy positions and returns
df['Position'] = np.where(df['MACD'] > df['Signal_Line'], 1, -1)
df['Strategy_Returns'] = df['Position'].shift(1) * df['daily_returns']
strategy_metrics = calculate_performance_metrics(df['close'] * (1 + df['Strategy_Returns'].cumsum()), "Strategy")

# Format and display results
def format_metric(x):
    if pd.isna(x):
        return "N/A"
    else:
        return f"{x:.2%}"

print("\nPerformance Analysis:")
print("\nMarket Performance:")
print(market_metrics.apply(format_metric))
print("\nStrategy Performance:")
print(strategy_metrics.apply(format_metric))

# Plot cumulative returns for comparison
plt.figure(figsize=(12, 6))
plt.plot((1 + calculate_clean_returns(df['close'])).cumprod(), label='Market')
plt.plot((1 + df['Strategy_Returns']).cumprod(), label='Strategy')
plt.title('Cumulative Returns Comparison')
plt.xlabel('Date')
plt.ylabel('Cumulative Return')
plt.legend()
plt.grid(True)
plt.show()
````

#### Analysis of Enhanced Bitcoin Performance Metrics

````python
# Add these helper functions first
def calculate_streak(series):
    """Calculate the longest streak of True values in a boolean series"""
    streak = 0
    max_streak = 0
    for value in series:
        if value:
            streak += 1
            max_streak = max(max_streak, streak)
        else:
            streak = 0
    return max_streak

def calculate_recovery_periods(drawdown):
    """Calculate the number of days to recover from drawdowns"""
    is_recovery = drawdown == 0
    recovery_periods = is_recovery.value_counts()
    return recovery_periods.get(True, 0)

# Update the enhanced performance metrics function
def enhanced_performance_metrics(prices, name="", annualization_factor=252):
    try:
        # Calculate clean returns
        returns = calculate_clean_returns(prices)
        
        # Basic return calculations
        total_return = (prices.iloc[-1] / prices.iloc[0]) - 1
        n_years = len(returns) / annualization_factor
        annual_return = (1 + total_return) ** (1/n_years) - 1
        
        # Rolling metrics
        rolling_std = returns.rolling(window=30).std()
        volatility = rolling_std.mean() * np.sqrt(annualization_factor)
        
        # Risk-adjusted metrics
        excess_returns = returns - 0.02/annualization_factor
        sharpe_ratio = np.sqrt(annualization_factor) * (excess_returns.mean() / returns.std())
        
        # Drawdown calculation
        peak = prices.expanding(min_periods=1).max()
        drawdown = (prices - peak) / peak
        max_drawdown = drawdown.min()
        
        # Calculate additional metrics
        metrics = pd.Series({
            'Total Return': total_return,
            'Annual Return': annual_return,
            'Volatility': volatility,
            'Sharpe Ratio': sharpe_ratio,
            'Max Drawdown': max_drawdown,
            'Win Rate': len(returns[returns > 0]) / len(returns),
            'Avg Daily Return': returns.mean(),
            'Calmar Ratio': annual_return / abs(max_drawdown),
            'Daily VaR 95%': returns.quantile(0.05),
            'Recovery Days': calculate_recovery_periods(drawdown),
            'Longest Winning Streak': calculate_streak(returns > 0),
            'Longest Losing Streak': calculate_streak(returns < 0)
        })
        
        return metrics
    
    except Exception as e:
        print(f"Error in {name} calculations: {e}")
        return pd.Series({
            'Total Return': np.nan,
            'Annual Return': np.nan,
            'Volatility': np.nan,
            'Sharpe Ratio': np.nan,
            'Max Drawdown': np.nan,
            'Win Rate': np.nan,
            'Avg Daily Return': np.nan,
            'Calmar Ratio': np.nan,
            'Daily VaR 95%': np.nan,
            'Recovery Days': np.nan,
            'Longest Winning Streak': np.nan,
            'Longest Losing Streak': np.nan
        })

# Update the format_metrics function to handle new metrics
def format_metrics(metrics):
    formatted = {}
    for k, v in metrics.items():
        if pd.isna(v):
            formatted[k] = "N/A"
        elif k in ['Total Return', 'Annual Return', 'Volatility', 'Max Drawdown', 'Win Rate', 'Daily VaR 95%']:
            formatted[k] = f"{v:.2%}"
        elif k in ['Sharpe Ratio', 'Calmar Ratio']:
            formatted[k] = f"{v:.2f}"
        elif k in ['Recovery Days', 'Longest Winning Streak', 'Longest Losing Streak']:
            formatted[k] = f"{int(v)}"
        else:
            formatted[k] = f"{v:.4f}"
    return pd.Series(formatted)

# Calculate and display enhanced metrics
enhanced_metrics = enhanced_performance_metrics(df['close'], "Market")
print("\nEnhanced Market Analysis:")
print(format_metrics(enhanced_metrics))
````


#### **Market Regime Analysis and Advanced Pattern Detection(BULL/BEAR)**
**Let's add market regime analysis to identify different market states (bull/bear markets) and detect important patterns. Here's the code:**


````python
def analyze_market_regimes(prices, returns, window=30):
    """Analyze market regimes and patterns"""
    
    # Calculate trend indicators
    sma_50 = prices.rolling(window=50).mean()
    sma_200 = prices.rolling(window=200).mean()
    
    # Define market regimes
    bull_market = sma_50 > sma_200
    
    # Calculate regime statistics
    volatility = returns.rolling(window=window).std() * np.sqrt(252)
    regime_stats = pd.DataFrame({
        'Price': prices,
        'Returns': returns,
        'Volatility': volatility,
        'Regime': np.where(bull_market, 'Bull', 'Bear')
    })
    
    # Plot market regimes
    plt.figure(figsize=(15, 10))
    
    # Subplot 1: Price with Moving Averages
    plt.subplot(2, 1, 1)
    plt.plot(prices.index, prices, label='Price', alpha=0.7)
    plt.plot(sma_50.index, sma_50, label='50-day MA', alpha=0.7)
    plt.plot(sma_200.index, sma_200, label='200-day MA', alpha=0.7)
    plt.fill_between(prices.index, 0, prices.max(), 
                     where=bull_market,
                     color='green', alpha=0.1)
    plt.fill_between(prices.index, 0, prices.max(), 
                     where=~bull_market,
                     color='red', alpha=0.1)
    plt.title('Bitcoin Market Regimes')
    plt.legend()
    
    # Subplot 2: Regime Statistics
    plt.subplot(2, 1, 2)
    for regime in ['Bull', 'Bear']:
        regime_data = returns[regime_stats['Regime'] == regime]
        plt.hist(regime_data, bins=50, alpha=0.5, 
                label=f'{regime} Market', 
                density=True)
    plt.title('Return Distribution by Market Regime')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Calculate regime statistics
    regime_summary = pd.DataFrame({
        'Bull Market': {
            'Days': bull_market.sum(),
            'Avg Return': returns[bull_market].mean() * 252,
            'Volatility': returns[bull_market].std() * np.sqrt(252),
            'Max Drawdown': (prices[bull_market] / prices[bull_market].expanding().max() - 1).min()
        },
        'Bear Market': {
            'Days': (~bull_market).sum(),
            'Avg Return': returns[~bull_market].mean() * 252,
            'Volatility': returns[~bull_market].std() * np.sqrt(252),
            'Max Drawdown': (prices[~bull_market] / prices[~bull_market].expanding().max() - 1).min()
        }
    }).T
    
    print("\nMarket Regime Analysis:")
    print(regime_summary.round(4))

# Run the analysis
returns = calculate_clean_returns(df['close'])
analyze_market_regimes(df['close'], returns)
````

````python
def detect_technical_patterns(prices, window=20):
    """Detect common technical patterns in price data"""
    
    df_patterns = pd.DataFrame(index=prices.index)
    
    # Calculate required indicators
    high = prices.rolling(window=window).max()
    low = prices.rolling(window=window).min()
    sma_20 = prices.rolling(window=20).mean()
    sma_50 = prices.rolling(window=50).mean()
    
    # Pattern 1: Double Top
    double_top = ((prices > high.shift(1)) & 
                  (prices.shift(-window) < prices) & 
                  (prices.shift(-1) < prices))
    
    # Pattern 2: Double Bottom
    double_bottom = ((prices < low.shift(1)) & 
                    (prices.shift(-window) > prices) & 
                    (prices.shift(-1) > prices))
    
    # Pattern 3: Support/Resistance Levels
    support = low.rolling(window=window).mean()
    resistance = high.rolling(window=window).mean()
    
    # Pattern 4: Golden Cross / Death Cross
    golden_cross = (sma_20 > sma_50) & (sma_20.shift(1) <= sma_50.shift(1))
    death_cross = (sma_20 < sma_50) & (sma_20.shift(1) >= sma_50.shift(1))
    
    # Visualize patterns
    plt.figure(figsize=(15, 12))
    
    # Price and Moving Averages
    plt.subplot(3, 1, 1)
    plt.plot(prices.index, prices, label='Price', alpha=0.7)
    plt.plot(sma_20.index, sma_20, label='20-day MA', alpha=0.7)
    plt.plot(sma_50.index, sma_50, label='50-day MA', alpha=0.7)
    
    # Mark Golden/Death Crosses
    plt.scatter(prices.index[golden_cross], 
               prices[golden_cross], 
               marker='^', color='g', s=100, 
               label='Golden Cross')
    plt.scatter(prices.index[death_cross], 
               prices[death_cross], 
               marker='v', color='r', s=100, 
               label='Death Cross')
    plt.title('Bitcoin Price with Technical Patterns')
    plt.legend()
    
    # Support/Resistance
    plt.subplot(3, 1, 2)
    plt.plot(prices.index, prices, label='Price', alpha=0.7)
    plt.plot(support.index, support, 'g--', label='Support', alpha=0.5)
    plt.plot(resistance.index, resistance, 'r--', label='Resistance', alpha=0.5)
    plt.title('Support and Resistance Levels')
    plt.legend()
    
    # Double Tops/Bottoms
    plt.subplot(3, 1, 3)
    plt.plot(prices.index, prices, label='Price', alpha=0.7)
    plt.scatter(prices.index[double_top], 
               prices[double_top], 
               marker='v', color='r', s=100, 
               label='Double Top')
    plt.scatter(prices.index[double_bottom], 
               prices[double_bottom], 
               marker='^', color='g', s=100, 
               label='Double Bottom')
    plt.title('Double Tops and Bottoms')
    plt.legend()
    
    plt.tight_layout()
    plt.show()
    
    # Calculate pattern statistics
    pattern_stats = pd.DataFrame({
        'Pattern': ['Golden Cross', 'Death Cross', 'Double Top', 'Double Bottom'],
        'Count': [golden_cross.sum(), death_cross.sum(), 
                 double_top.sum(), double_bottom.sum()],
        'Avg Return (30d)': [
            prices[golden_cross].pct_change(30).mean(),
            prices[death_cross].pct_change(30).mean(),
            prices[double_top].pct_change(30).mean(),
            prices[double_bottom].pct_change(30).mean()
        ]
    })
    
    print("\nPattern Analysis Results:")
    print(pattern_stats.round(4))
    
    return pattern_stats

# Run pattern detection
pattern_results = detect_technical_patterns(df['close'])
````

#### Volume Analysis & Price Volume Relationships
##### *for validating price patterns and market movements*

````python
def analyze_volume_patterns(df, window=20):
    """Analyze volume patterns and their relationship with price movements"""
    
    # Calculate volume indicators
    df_vol = pd.DataFrame(index=df.index)
    df_vol['volume'] = df['volume']
    df_vol['volume_sma'] = df['volume'].rolling(window=window).mean()
    df_vol['volume_std'] = df['volume'].rolling(window=window).std()
    df_vol['price'] = df['close']
    
    # Volume trends and anomalies
    df_vol['volume_z_score'] = (df_vol['volume'] - df_vol['volume_sma']) / df_vol['volume_std']
    df_vol['high_volume'] = df_vol['volume_z_score'] > 2
    
    # Price-volume relationship
    df_vol['price_change'] = df_vol['price'].pct_change()
    df_vol['volume_price_corr'] = df_vol['price'].rolling(window).corr(df_vol['volume'])
    
    # Visualization
    fig, axes = plt.subplots(3, 1, figsize=(15, 12))
    
    # Price and Volume
    ax1 = axes[0]
    ax1.plot(df_vol.index, df_vol['price'], label='Price', color='blue')
    ax1.set_title('Bitcoin Price and Volume Analysis')
    ax1.set_ylabel('Price (USD)')
    ax1.legend(loc='upper left')
    
    # Volume bars with highlights
    ax1v = ax1.twinx()
    ax1v.bar(df_vol.index, df_vol['volume'], 
             color=np.where(df_vol['high_volume'], 'red', 'gray'),
             alpha=0.3, label='Volume')
    ax1v.plot(df_vol.index, df_vol['volume_sma'], 
              color='orange', label='Volume SMA')
    ax1v.set_ylabel('Volume')
    ax1v.legend(loc='upper right')
    
    # Volume Z-Score
    axes[1].plot(df_vol.index, df_vol['volume_z_score'])
    axes[1].axhline(y=2, color='r', linestyle='--')
    axes[1].axhline(y=-2, color='r', linestyle='--')
    axes[1].set_title('Volume Z-Score')
    
    # Price-Volume Correlation
    axes[2].plot(df_vol.index, df_vol['volume_price_corr'])
    axes[2].axhline(y=0, color='k', linestyle='--')
    axes[2].set_title('Price-Volume Rolling Correlation')
    
    plt.tight_layout()
    plt.show()
    
    # Calculate volume statistics
    volume_stats = pd.DataFrame({
        'Metric': [
            'Average Daily Volume',
            'Volume Volatility',
            'High Volume Days %',
            'Price-Volume Correlation',
            'Up Days with High Volume %'
        ],
        'Value': [
            df_vol['volume'].mean(),
            df_vol['volume'].std() / df_vol['volume'].mean(),
            (df_vol['high_volume'].sum() / len(df_vol)) * 100,
            df_vol['price'].corr(df_vol['volume']),
            (df_vol['high_volume'] & (df_vol['price_change'] > 0)).sum() / df_vol['high_volume'].sum() * 100
        ]
    })
    
    print("\nVolume Analysis Statistics:")
    print(volume_stats.round(2))
    
    return df_vol

# Run volume analysis
volume_analysis = analyze_volume_patterns(df)
````

### ADDING MOMENTUM INDICATORS

````python
def analyze_volume_momentum(df_vol, window=14):
    """Add momentum analysis to volume patterns"""
    
    # Calculate momentum indicators
    df_momentum = df_vol.copy()
    
    # Money Flow Index (MFI)
    typical_price = (df['high'] + df['low'] + df['close']) / 3
    money_flow = typical_price * df['volume']
    
    positive_flow = pd.Series(0, index=df.index)
    negative_flow = pd.Series(0, index=df.index)
    
    # Calculate positive and negative money flow
    positive_flow[typical_price > typical_price.shift(1)] = money_flow[typical_price > typical_price.shift(1)]
    negative_flow[typical_price < typical_price.shift(1)] = money_flow[typical_price < typical_price.shift(1)]
    
    positive_mf = positive_flow.rolling(window=window).sum()
    negative_mf = negative_flow.rolling(window=window).sum()
    
    mfi = 100 - (100 / (1 + positive_mf / negative_mf))
    
    # Add to dataframe
    df_momentum['MFI'] = mfi
    
    # Visualization
    plt.figure(figsize=(15, 8))
    
    # Plot MFI
    plt.subplot(2, 1, 1)
    plt.plot(df_momentum.index, df_momentum['MFI'], label='MFI')
    plt.axhline(y=80, color='r', linestyle='--')
    plt.axhline(y=20, color='g', linestyle='--')
    plt.title('Money Flow Index')
    plt.legend()
    
    # Print momentum statistics
    momentum_stats = pd.DataFrame({
        'Metric': [
            'Overbought Days (MFI > 80)',
            'Oversold Days (MFI < 20)',
            'Average MFI',
            'MFI Volatility'
        ],
        'Value': [
            (df_momentum['MFI'] > 80).mean() * 100,
            (df_momentum['MFI'] < 20).mean() * 100,
            df_momentum['MFI'].mean(),
            df_momentum['MFI'].std()
        ]
    })
    
    print("\nMomentum Analysis:")
    print(momentum_stats.round(2))
    
    return df_momentum

# Run momentum analysis
momentum_analysis = analyze_volume_momentum(volume_analysis)
````
