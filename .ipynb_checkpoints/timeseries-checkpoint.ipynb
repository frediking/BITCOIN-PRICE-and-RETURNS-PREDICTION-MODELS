```python
import pandas as pd
import numpy as np  
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import r2_score
```

```python
df = pd.read_csv('Bitcoin_Historical_Data.csv', parse_dates=['Date'])
df.sort_values("Date", inplace=True)
```

```python
df.head()
```

```python
#TARGET AND ROLLING LAG FEATURES

# target shift
df['Close_Tomorrow'] = df['Close'].shift(-1)
```

```python
# lag features
df['close_lag_1'] = df['Close'].shift(1)
df['close_lag_3'] = df['Close'].shift(3)
df['close_lag_7'] = df['Close'].shift(7)
```

```python
df['rolling_max_7'] = df['High'].rolling(window=7).max()
df['rolling_mean_7'] = df['Close'].rolling(window=7).mean()
```

```python
df.dropna(inplace=True)
```

```python
X = ['close_lag_1', 'close_lag_3', 'close_lag_7', 'rolling_max_7', 'rolling_mean_7']
y = 'Close_Tomorrow'
```

```python
df.head()
```

```python
train = df[df["Date"] < "2021-01-01 00:00:00+00:00"]
test = df[df["Date"] >= "2021-01-01 00:00:00+00:00"]
```

```python
X_train = train[X]
y_train = train[y]
X_test = test[X]
y_test = test[y]
```

```python
rfrmodel = RandomForestRegressor(n_estimators=100, random_state=42)
rfrmodel.fit(X_train, y_train)
```

```python
Y_pred = rfrmodel.predict(X_test)
```

```python
TESTr2 = r2_score(y_test, Y_pred)
TESTr2
```
